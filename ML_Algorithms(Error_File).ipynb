```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv('spam.csv')
df.head()
```

```python
df['Category']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Category',axis='columns'),df.Category,test_size=0.2)
```

```python
x_train
```

##Using the count vectorizer to identify each messsage separately

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
x_train_converted=v.fit_transform(x_train.Message)
x_train_converted.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_converted,y_train)
```

```python
x_test_converted=v.transform(x_test.Message)
model.score(x_test_converted,y_test)
```

```python
y_pred=model.predict(x_test_converted)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
email_converted=v.transform(emails)
model.predict(email_converted)
```

```python
from sklearn.pipeline import Pipeline
preprocessor=ColumnTransformer(
    transformers=[
        ('vectorizer',CountVectorizer(),['Message'])
    ],
    remainder='passthrough'
)
pipe=Pipeline([
    ('preprocessor',preprocessor),
    ('model',MultinomialNB())
])

pipe.fit(x_train,y_train)
pipe.predict(x_test)
```
