```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data-export.csv")
df.head()
```

```python
df.columns=df.iloc[0]
df.head()
```

```python
df=df.drop(index=0).reset_index(drop=True)
df.head()
```

```python
df.info()
```

#Changing the column **names**

```python
df.columns=["ChannelGroup","DateTime","Users","Sessions","Engaged_sessions","Average_engagement_time_per_session","Engaged_sessions_per_user","Events_per_session","Engagement_rate","Event_count"]
df.head()
```

# Let's convert datehour to the required format

```python
df['DateTime']=pd.to_datetime(df['DateTime'],format="%Y%m%d%H")
df.head()
```

#Let's convert to int columns along with double columns

```python
numeric_cols=df.columns.drop(["ChannelGroup","DateTime"])
print(numeric_cols)
```

```python
df[numeric_cols]=df[numeric_cols].apply(pd.to_numeric,errors='coerce')
df.head()
```

```python
df.info()
```

```python
df['Hour']=df["DateTime"].dt.hour
```

```python
df.info()
```

# **# ###So data cleaning is done**

```python
df.describe()
```

# What patterns or trends can you observe in website sessions or users over time?

```python
start = df['Users'].iloc[0]   # first value in time
end = df['Users'].iloc[-1]    # last value in time

change_percent = ((end - start) / start) * 100
print(f"{change_percent:.2f}% growth was shown in users")
```

```python
df_non_zero=df[df['Users']>0]
df_non_zero.head()
df_non_zero.tail()

start=df_non_zero['Users'].iloc[0]
end=df_non_zero['Users'].iloc[-1]
change_percent=((end-start)/start)*100
print(f"{change_percent:.2f}% growth was shown in users")
```

```python
max_users=df['Users'].max()
max_row=df[df['Users']==max_users]

max_dates=max_row['DateTime'].dt.date.unique()
print(f"{max_users} were found in this {max_dates}")
```

```python
df["Date"]=df["DateTime"].dt.date

daily_users=df.groupby("Date")["Users"].sum()
print(daily_users)

max_users=daily_users.max()
max_date=daily_users.idxmax()

print(f"{max_users} occurred on this {max_date}")
```

```python
print(f"{max_users} users occurred on this {max_date}")
```

```python
daily_sessions=df.groupby("Date")["Sessions"].sum()
print(daily_sessions)

max_session=daily_sessions.max()
max_session_date=daily_sessions.idxmax()

print(f"{max_session} session occurred on this date {max_session_date}")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12,6))
df.sort_values("DateTime")
df.groupby("DateTime")[["Users","Sessions"]].sum().plot(ax=plt.gca())
plt.title("Users and Sessions Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.show()
```

# ##Which marketing channel brought the highest number of users?

```python
highest_user_channel=df.groupby("ChannelGroup")["Users"].sum().reset_index()
print(highest_user_channel)
```

```python
max_users_of_channel=highest_user_channel['Users'].max()


max_idx=highest_user_channel['Users'].idxmax()


max_user_channel=highest_user_channel.loc[max_idx,'ChannelGroup']

print(f"{max_users_of_channel} users traffic came from this channel {max_user_channel}")

```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="ChannelGroup",y="Users",data=highest_user_channel,palette='viridis')

plt.title("Users by Channel")
plt.xlabel("Channels")
plt.ylabel("Users")
plt.xticks(rotation=45)
plt.show()
```

# ##Which channel has highest average engagement time?

```python
df.head()
```

```python
highest_average_engagement=df.groupby("ChannelGroup")["Average_engagement_time_per_session"].max().reset_index()
print(highest_average_engagement)
```

```python
highest_average_session_time=highest_average_engagement["Average_engagement_time_per_session"].max()
max_idx_highest_average_session_time=highest_average_engagement["Average_engagement_time_per_session"].idxmax()
highest_average_session_time_channel=highest_average_engagement.loc[max_idx_highest_average_session_time,"ChannelGroup"]

print(f"{highest_average_session_time} time was found for {highest_average_session_time_channel}")
```

# **##How does engagement rate vary accross different channels? **

```python
average_engagement_rate_perchannel=df.groupby("ChannelGroup")["Engagement_rate"].mean().reset_index()
print(average_engagement_rate_perchannel)
```

# ##Visualize both engaged sessions and non-engaged sessions per channel

```python
session_df=df.groupby("ChannelGroup")[["Sessions","Engaged_sessions"]].sum().reset_index()
session_df["Non_Engaged_Sessions"]=session_df["Sessions"]-session_df["Engaged_sessions"]
session_df.head()
```

```python
session_melted=session_df.melt(
    id_vars="ChannelGroup",
    value_vars=["Engaged_sessions","Non_Engaged_Sessions"],

)

```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="ChannelGroup",
    y="value",
    hue="variable",
    data=session_melted,
    palette="magma")
plt.title("Engaged vs Non-Engaged Sessions per Channel")
plt.xlabel("Channel")
plt.ylabel("Number of Sessions")
plt.xticks(rotation=45)
plt.legend(title="Session Type")
plt.show()
```

# ##At what hour of the day for each channel how much was the traffic?

```python
traffic_matrix = df.pivot_table(
    index="ChannelGroup",
    columns="Hour",
    values="Users",
    aggfunc="sum",
    fill_value=0
)
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(traffic_matrix, cmap="YlOrRd")
plt.title("Traffic (Users) by Hour and Channel")
plt.xlabel("Hour of Day")
plt.ylabel("Channel")
plt.show()
```
